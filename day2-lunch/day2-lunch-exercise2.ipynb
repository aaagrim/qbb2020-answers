# Making dictionary of mapping file

```python
mapping_file = open('identify.txt', 'r')
                
mapping_dict = {}
        
for line in mapping_file:
    fields = line.rstrip().split('\t')
    mapping_dict[fields[0]] = fields[1]
    
mapping_file.close()
```

# Defining the function

```python
def id_map(mapping, experimental, NA = True ):
        mapping_file = open(mapping, 'r')
        experimental_file = open(experimental, 'r')
        translation = open('translated.txt', 'w')
        
        #writing the new header
        translation.write('t_id' + '\t' + 'chr' + '\t' + 'strand' + '\t' + 'start' + '\t' + 'end' + '\t' +  't_name' + '\t' +
                          'num_exons' + '\t' + 'length' + '\t' + 'uniprot_id' + '\t' + 'gene_name' + '\t' + 'cov'+ '\t' + 'FPKM' + '\n')
        
        #Create a dictionary out of the mapping file
        mapping_dict = {}
        
        for line in mapping_file:
            fields = line.rstrip().split('\t')
            mapping_dict[fields[0]] = fields[1]
        
        #parse through the experimental file
        for i, line in enumerate(experimental_file):
            if i == 0:
                continue
            if i >= 100:
                break
    
            file_list = line.rstrip().split()
            FB_ID = file_list[8]
    
    
            #Determine if the FB_ID is located in the mapping dictionary and write new line
            if FB_ID in mapping_dict.keys():
                a = mapping_dict[FB_ID]
                mapping = '{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n'.format(file_list[0],file_list[1],file_list[2],
                                                                                    file_list[3],file_list[4],file_list[5],
                                                                                   file_list[6],file_list[7], a, file_list[9],
                                                                                   file_list[10],file_list[11])
                translation.write(mapping)
            
            else:
                if NA == True:
                    b = 'NA'
                    mapping2 = '{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n'.format(file_list[0],file_list[1],file_list[2],
                                                                                    file_list[3],file_list[4],file_list[5],
                                                                                   file_list[6],file_list[7], b, file_list[9],
                                                                                   file_list[10],file_list[11])
                    translation.write(mapping2)

        mapping_file.close()
        experimental_file.close()
        translation.close()
```

# Run function

```python
id_map('identify.txt','t_data.ctab') #should input False as 3rd argument if wanting to skip empty lines
```
