# Comparing the E4 and E5 files

```python
#Using numpy to read in the file as an array using start position and methylation score
E4_sites = np.genfromtxt("SRR3083926_1.chr6_bismark_bt2_pe.sorted.bedGraph", usecols=(1,3), skip_header=1)
methylation_change = []

for i in range(0, len(E4_sites)):
    if E4_sites[i, 1] == 0:#skip unscored positions
        continue
    else:
        methylation_change.append([int(E4_sites[i, 0]), E4_sites[i, 1]])
    
print(methylation_change)
```

```python
#Using numpy to read in the file as an array using start position and methylation score
E5_sites = np.genfromtxt("SRR3083929_1.chr6_bismark_bt2_pe.sorted.bedGraph", usecols=(1,3), skip_header=1)

E4_index = 0
E5_index = 0
meth_folds = []#Use a list of lists to keep track of position and the scores for E4 and E5 at that position. 
while (E4_index < len(methylation_change)):
    if methylation_change[E4_index][0] == E5_sites[E5_index][0]:
        fold_change = E5_sites[E5_index][1] / methylation_change[E4_index][1]
        
        if fold_change == 0:#skip unscored E5 positions
            E4_index += 1
            E5_index += 1
        else:
            meth_folds.append([methylation_change[E4_index][0],E5_sites[E5_index][1],methylation_change[E4_index][1]])
        
            E4_index += 1
            E5_index += 1
        
    elif methylation_change[E4_index][0] < E5_sites[E5_index][0]:
        E4_index += 1
    
    else:
        E5_index += 1
        
```

```python
#Making a numpy array from the reference bed file, containing positons and gene name
gene_conv = np.genfromtxt("mm10_refseq_genes_chr6_50M_60M.bed", usecols=(4,5, 12), dtype = str)
gene_conv
```

```python
#generating a dictionary with the gene names and the various coordinates associated with the genes
gene_dic = {}
for i in range(0, len(gene_conv)):
    gene_dic.setdefault(gene_conv[i][2],[])
    gene_dic[gene_conv[i][2]].append((int(gene_conv[i][0]), int(gene_conv[i][1])))

print(gene_dic)
```

```python
average_fold_change = {}
for key, item in gene_dic.items():
    average_fold_change.setdefault(key, [0.0,0.0,0])#intializing the dictionary
    for i in range(0, len(meth_folds)):
        for it in item:
            if it[0] <= meth_folds[i][0] and meth_folds[i][0] <= it[1]:#check if the position occurs between two coordinates
                average_fold_change[key][0] += meth_folds[i][1]#sum of metyhlation scores for E5
                average_fold_change[key][1] += meth_folds[i][2]#sum of methylation scores for E4
                average_fold_change[key][2] += 1 #the count of each site - used for averaging the sums
                break
    if average_fold_change[key][2] == 0:#removing an genes without methylation scores
        del average_fold_change[key]
            
print(average_fold_change)
```

```python
gene_final = []
for key, item in average_fold_change.items():
    avg_E5 = item[0]/item[2]
    avg_E4 = item[1]/item[2]
    fold = avg_E5/avg_E4 #Calculating the fold change 
    gene_final.append([key,fold])
```

```python
gene_final
np.savetxt("gene_methylation.txt", gene_final, fmt="%s", delimiter="\t")#saving our data to a txt file using numpy
```
